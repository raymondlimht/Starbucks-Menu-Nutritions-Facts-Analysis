## Data Cleaning

```python
# check total fat content unique value

menu["TotalFat(g)"].unique()
```

Found out '3 2' which is uncommon. The number shall be 3.2g

```python
menu[menu["TotalFat(g)"] == '3 2']
menu.loc[230:240]
```

```python
# replace "3 2" to "3.2"
menu.loc[237,"TotalFat(g)"]= "3.2"
```

```python
# Convert columns to numeric
menu["TotalFat(g)"] = pd.to_numeric(menu["TotalFat(g)"], errors='coerce')
```

```python
pd.set_option('display.max_columns', None)
```

```python
# convert VitaminA(%DV), VitaminC(%DV), Calcium(%DV), Iron(%DV) into float.

#remove "%", convert string to float.
columns = ["VitaminA(%DV)","VitaminC(%DV)", "Calcium(%DV)", "Iron(%DV)"] 

for column in columns:
    menu[column] = menu[column].str.rstrip("%").astype(float)/100
```

```python
menu.info()
```

### Dealing with 'Varies' in caffeine(mg)

```python
menu['Caffeine(mg)'].unique()
```

```python
# Convert columns to numeric
menu["Caffeine(mg)"] = pd.to_numeric(menu["Caffeine(mg)"], errors='coerce')
```

```python
menu["Caffeine(mg)"].unique()
```

```python
## Replace null with mean
menu['Caffeine(mg)'].fillna((menu['Caffeine(mg)'].mean()), inplace = True)
```

## 1. Drinks with the highest calories from the dataset




```python
menu.sort_values("Calories", ascending=False)
```

```python
# plot Beverage average calories


plt.figure(figsize=(25,8))
plot_order = menu.groupby('Beverage')['Calories'].mean().sort_values(ascending=False).index.values

ax = sns.barplot(x="Beverage", y="Calories", data = menu, ci=None,  order=plot_order,palette='BrBG_r')
plt.xticks(rotation=40, ha='right')
plt.title("Starbucks Beverage Average Calories", fontsize=25)
ax.bar_label(ax.containers[0])
plt.xlabel('Beverage',fontsize=17)
plt.ylabel('Average Calories', fontsize=17)
plt.show()
```

Comment: White Chocolate Mocha has the highest average calories, followed by Java Chip and banana chocolate smoothie. Chocolate beverages tend to have higher calories contents. Non-sweeten drinks such as Americano, Espresso, Brewed Coffee and Tazo Tea has the lower calories contents which are ideals for customers seeking for lower calories intake at the same time needed some caffeine boost. 

The high calories contents of the drinks could be coming from sugar and fat which usually has high calories. One could study the correlation between calories vs sugars/fat to find out more insights.

###  9.1 Probability distribution plot of Calories between 'Classic Espresso Drinks', 'Signature Espresso Drinks'

```python
categories = ['Classic Espresso Drinks', 'Signature Espresso Drinks']
```

```python

plt.figure(figsize=(8,5))
for category in categories:
    sns.kdeplot(menu.loc[(menu['Beverage_category']==category), 'Calories'],label= category)
plt.title('Calories distributions Classic Espresso vs Signature Espresso', size= 15)
plt.xlabel('Calories')
plt.ylabel('Probability Density')
plt.legend()
```

KDE plot is useful to compare the features side by sides. Based on the KDE distribution plot, 'Classic Espresso Drinks' distribution is higher around 100 Cals. On the other hands, Signature Espresso Drinks has higher calories around 180-250 cals.

```python
# classic & signautre coffee calories

figure, ax = plt.subplots(1,2, figsize=(12,6))
figure.suptitle('Signature vs Classic Espresso Drinks Calories Distribution Plot',fontsize=20)

signature = menu['Calories'].loc[menu['Beverage_category'] == 'Signature Espresso Drinks']
sns.distplot(ax=ax[0],x=signature, norm_hist=False, color=sns.color_palette("Paired")[1])
ax[0].set_title('Signature Espresso Drinks')

classic = menu['Calories'].loc[menu['Beverage_category'] == 'Classic Espresso Drinks']
sns.distplot(ax = ax[1], x = classic, norm_hist=False, color=sns.color_palette("Paired")[2])
ax[1].set_title('Classic Espresso Drinks')
```

### 9.2 Visualize the probability distributions of calories against other features 

By using Kdeplot, we are able to further analyse the distribution of beverage category between two features. 

```python
menu['Beverage_category'].unique()
```

```python
menu.columns.unique()
```

```python
# Setting up the beverage category
signature = menu.query("Beverage_category=='Signature Espresso Drinks'")
classic = menu.query("Beverage_category=='Classic Espresso Drinks'")
frapp = menu.query("Beverage_category=='Frappuccino® Blended Coffee'")
iced_shaken = menu.query("Beverage_category=='Shaken Iced Beverages'")
smoothie = menu.query("Beverage_category=='Smoothies'")
tazo = menu.query("Beverage_category=='Tazo® Tea Drinks'")
```

```python
## Plotting the KDE Plot

plt.figure(figsize=(8,8))
sns.kdeplot(data= smoothie, x='Calories', y='Protein(g)', 
            shade=True, label='Smoothie', cmap="Blues")

sns.kdeplot(data= iced_shaken, x='Calories', y='Protein(g)',
            shade=True, label='Iced Shaken', cmap="Greens")

handles = [mpatches.Patch(facecolor=plt.cm.Blues(200), label="Smoothie"),
           mpatches.Patch(facecolor=plt.cm.Greens(200), label="Iced Shaken")]
plt.legend(handles=handles, loc=2)
plt.title("Smoothie vs Iced Shaken probability plot", size=15)
```

We can see that "smoothies" generally has higher protein and calories content compared to Iced Shaken drinks.

```python
# KDE plot of Tazo Tea vs Frappuccino

plt.figure(figsize=(8,8))
sns.kdeplot(data= frapp, x='Calories', y='Protein(g)', 
            shade=True, label='Frappuccino', cmap="Reds")

sns.kdeplot(data= tazo, x='Calories', y='Protein(g)',
            shade=True, label='Tazo Tea', cmap="Blues")

handles = [mpatches.Patch(facecolor=plt.cm.Reds(200), label="Frappuccino"),
           mpatches.Patch(facecolor=plt.cm.Blues(200), label="Tazo Tea")]
plt.legend(handles=handles, loc=2)
plt.title("Tazo Tea vs Frappuccino probability plot", size=15)
```

```python
# KDE plot of Classic Coffee vs Smoothie 

plt.figure(figsize=(8,8))
sns.kdeplot(data= classic, x='Calories', y='Protein(g)', 
            shade=True, label='Classic coffee', cmap="Greens")

sns.kdeplot(data= smoothie, x='Calories', y='Protein(g)',
            shade=True, label='Smoothie', cmap="Reds")

handles = [mpatches.Patch(facecolor=plt.cm.Greens(200), label="Classic Coffee"),
           mpatches.Patch(facecolor=plt.cm.Reds(200), label="Smoothie")]
plt.legend(handles=handles, loc=2)
plt.title("Classic Coffee vs Smoothie probability plot", size=15)
```
